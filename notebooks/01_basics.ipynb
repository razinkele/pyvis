# PyVis Basics — Creating Interactive Networks

This notebook covers the fundamentals of creating interactive network visualizations with PyVis.

**What you'll learn:**
- Creating networks and adding nodes/edges
- Styling nodes with colors, shapes, and sizes
- Batch operations for adding multiple nodes
- Toggling physics simulation
- Using filter and select menus
- Exploring the vis.js options UI

```python
from pyvis.network import Network
```

## 1. Your First Network

Create a simple 3-node network with styled edges.

```python
net = Network(notebook=True, cdn_resources="in_line", height="500px")
net.add_node(1, label="Alice", color="#6366f1", shape="dot", size=25)
net.add_node(2, label="Bob", color="#22d3ee", shape="dot", size=25)
net.add_node(3, label="Charlie", color="#34d399", shape="dot", size=25)
net.add_edge(1, 2, width=2, title="friends")
net.add_edge(2, 3, width=1, title="colleagues")
net.add_edge(1, 3, width=3, title="family")
net.show("01_basic.html")
```

## 2. Batch Operations

Use `add_nodes()` and `add_edges()` to add multiple elements at once.

```python
net = Network(notebook=True, cdn_resources="in_line", height="500px")
net.add_nodes(
    [1, 2, 3, 4, 5],
    label=["A", "B", "C", "D", "E"],
    color=["#f87171", "#fbbf24", "#34d399", "#22d3ee", "#818cf8"],
    size=[30, 25, 20, 25, 30],
    title=["Node A", "Node B", "Node C", "Node D", "Node E"],
)
net.add_edges([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)])
net.show("01_batch.html")
```

## 3. Physics Simulation

Toggle the physics engine on and off. With physics disabled, nodes stay where they are.

```python
net.toggle_physics(False)
net.show("01_no_physics.html")
```

## 4. Node Shapes

vis.js supports many node shapes: `dot`, `star`, `triangle`, `square`, `diamond`, `triangleDown`, `hexagon`, `box`, `ellipse`, `circle`, `text`, and more.

```python
net = Network(notebook=True, cdn_resources="in_line", height="500px")
shapes = ["dot", "star", "triangle", "square", "diamond", "triangleDown", "hexagon"]
for i, shape in enumerate(shapes):
    net.add_node(i, label=shape, shape=shape, size=25, color="#6366f1")
# Connect in a circle
for i in range(len(shapes) - 1):
    net.add_edge(i, i + 1)
net.add_edge(len(shapes) - 1, 0)
net.show("01_shapes.html")
```

## 5. Filter and Select Menus

Setting `select_menu=True` adds a dropdown to highlight nodes. Setting `filter_menu=True` adds controls to filter nodes by their properties.

```python
net = Network(notebook=True, cdn_resources="in_line", height="600px",
              select_menu=True, filter_menu=True)
for i in range(1, 16):
    group = "even" if i % 2 == 0 else "odd"
    color = "#22d3ee" if i % 2 == 0 else "#f87171"
    net.add_node(i, label=str(i), group=group, title=f"Node {i} ({group})", color=color)
for i in range(1, 15):
    net.add_edge(i, i + 1)
net.add_edge(15, 1)
net.show("01_filter.html")
```

## 6. Options UI

Use `show_buttons()` to expose the vis.js configuration UI directly below the graph. This lets you experiment with physics settings and export the JSON configuration.

```python
net.show_buttons(filter_="physics")
net.show("01_buttons.html")
```
