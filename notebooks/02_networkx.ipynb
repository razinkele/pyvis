# NetworkX Integration

PyVis can import graphs directly from [NetworkX](https://networkx.org/), Python's most popular graph library. The `from_nx()` method converts any NetworkX graph into an interactive visualization.

**What you'll learn:**
- Importing NetworkX graphs with `from_nx()`
- Custom node size and edge weight transforms
- Different graph generators (random tree, Karate Club, Barabasi-Albert)
- Adding hover data from the adjacency list

```python
import networkx as nx
from pyvis.network import Network
```

## 1. Random Tree

The simplest import — generate a random tree and visualize it.

```python
G = nx.random_labeled_tree(20, seed=42)
net = Network(notebook=True, cdn_resources="in_line", height="500px")
net.from_nx(G)
net.show("02_tree.html")
```

## 2. Karate Club Graph

Zachary's Karate Club is a classic social network dataset. Each node is a club member, edges represent social ties.

```python
G = nx.karate_club_graph()
net = Network(notebook=True, cdn_resources="in_line", height="500px",
              bgcolor="#1a1a2e", font_color="#e2e8f0")
net.from_nx(G)
net.show("02_karate.html")
```

## 3. Custom Transforms

`from_nx()` accepts `node_size_transf` and `edge_weight_transf` functions to control how node sizes and edge widths are calculated.

```python
G = nx.barabasi_albert_graph(30, 2, seed=42)
# Add weights based on node distance
for u, v in G.edges():
    G[u][v]["weight"] = abs(u - v)

net = Network(notebook=True, cdn_resources="in_line", height="500px")
net.from_nx(
    G,
    default_node_size=10,
    node_size_transf=lambda x: x * 2,
    edge_weight_transf=lambda w: max(1, w * 0.3),
)
net.show("02_transforms.html")
```

## 4. Neighbor Data on Hover

Use `get_adj_list()` to build hover tooltips showing each node's neighbors. Set `node["value"]` to the degree to make hub nodes larger.

```python
G = nx.watts_strogatz_graph(20, 4, 0.3, seed=42)
net = Network(notebook=True, cdn_resources="in_line", height="500px")
net.from_nx(G)

neighbor_map = net.get_adj_list()
for node in net.nodes:
    nid = node["id"]
    neighbors = neighbor_map[nid]
    node["title"] = f"Node {nid}<br>Degree: {len(neighbors)}<br>Neighbors: {', '.join(str(n) for n in neighbors)}"
    node["value"] = len(neighbors)

net.show("02_neighbors.html")
```

## 5. Scale-Free Network

Barabasi-Albert graphs exhibit preferential attachment — a few hub nodes accumulate many connections.

```python
G = nx.barabasi_albert_graph(50, 2, seed=7)
net = Network(notebook=True, cdn_resources="in_line", height="600px",
              bgcolor="#0f172a", font_color="#e2e8f0")
net.from_nx(G)

# Size nodes by degree
neighbor_map = net.get_adj_list()
for node in net.nodes:
    degree = len(neighbor_map[node["id"]])
    node["value"] = degree
    node["title"] = f"Node {node['id']}<br>Degree: {degree}"
    # Color hubs differently
    if degree > 5:
        node["color"] = "#f87171"
    elif degree > 3:
        node["color"] = "#fbbf24"
    else:
        node["color"] = "#6366f1"

net.barnes_hut(gravity=-3000)
net.show("02_scale_free.html")
```

## 6. Adding Custom Node Attributes from NetworkX

Node attributes set in NetworkX carry through to PyVis.

```python
G = nx.cycle_graph(10)
G.nodes[0]["title"] = "Start node"
G.nodes[0]["group"] = "special"
G.nodes[0]["color"] = "#f87171"
G.nodes[5]["title"] = "Halfway node"
G.nodes[5]["group"] = "special"
G.nodes[5]["color"] = "#34d399"
G.add_node(10, size=30, label="Hub", title="Central hub", color="#fbbf24")
G.add_edge(0, 10, weight=5)
G.add_edge(5, 10, weight=5)

net = Network(notebook=True, cdn_resources="in_line", height="500px")
net.from_nx(G)
net.show("02_attributes.html")
```
